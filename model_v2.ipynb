```python
from glob import glob
import yaml
import torch
```

```python
# 이미지들의 주소 리스트로 만들어줌
train_img_list = glob("/content/drive/MyDrive/De_Bugging.v2/train/images/*.jpg")
valid_img_list = glob("/content/drive/MyDrive/De_Bugging.v2/valid/images/*.jpg")
```

```python
# 리스트를 txt파일로 생성
with open("/content/drive/MyDrive/De_Bugging.v2/train.txt", "w") as f:
	f.write("\n".join(train_img_list) + "\n")
    
with open("/content/drive/MyDrive/De_Bugging.v2/valid.txt", "w") as f:
	f.write("\n".join(valid_img_list) + "\n")

with open("/content/drive/MyDrive/De_Bugging.v2/data.yaml", "r") as f:
	data = yaml.safe_load(f)
```

```python
data["train"] = "/content/drive/MyDrive/De_Bugging.v2/train.txt"
data["valid"] = "/content/drive/MyDrive/De_Bugging.v2/valid.txt"
 
with open("/content/drive/MyDrive/De_Bugging.v2/data.yaml", "w") as f:
	yaml.dump(data, f)
```

```python
# 파라미터 설정 (batch=128, epochs=100, yolov5s 모델 이용) 후 모델 학습.
# 구글 드라이브의 drive/MyDrive/De_Bugging.v2/yolov5/runs/train/model_v4 경로에 모델이 저장된다.

!python /content/drive/MyDrive/De_Bugging.v2/yolov5/train.py --img 320 --batch 128 --epochs 100 --data /content/drive/MyDrive/De_Bugging.v2/data.yaml --cfg /content/drive/MyDrive/De_Bugging.v2/yolov5/models/yolov5s.yaml --weights yolov5s.pt --name model_v4
```
